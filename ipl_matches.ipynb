```python
toss_win_match_win=ipl[ipl['toss_winner']==ipl['winner']]
toss_win_match_win.head()
```

```python
# find percentage to show which team won the match who is won toss
toss_winner_match_winner=np.sum(ipl['toss_winner']==ipl['winner'])
toss_losser_match_winner=np.sum(ipl['toss_winner']!=ipl['winner'])
total_match_conduct=ipl['id'].value_counts().sum()
toss_winner_match_winner=(toss_winner_match_winner/total_match_conduct)*100
toss_losser_match_winner=(toss_losser_match_winner/total_match_conduct)*100
print(f'toss winner match winner : {round(toss_winner_match_winner,2)}%')
print(f'toss losser match winner : {round(toss_losser_match_winner,2)}%')
```
