```python
#show all missing data from 'city' column
ipl[ipl['city'].isnull()]
```

```python
#mising values of 'city' column replace with 'UAE' coz,  Dubai is a country of UAE
ipl['city']=np.where((ipl['city'].isnull()),"UAE",ipl['city'])
```

```python
ipl.isnull().sum()
```

```python
ipl[ipl['winner'].isnull()]
```

```python
#mising values of 'winner' column replace with 'nowinner' coz, no winner
ipl['winner']=np.where((ipl['winner'].isnull()),"nowinner",ipl['winner'])
```

```python
ipl[ipl['winner'].isnull()==0][300:301]
```

```python
#mising values of 'palyerofthematch' column replace with 'Noplayerofthematch' coz, no one won
ipl['player_of_match']=np.where((ipl['player_of_match'].isnull()),"Noplayerofthematch",ipl['player_of_match'])
```

```python
ipl[ipl['umpire1'].isnull()]
```

```python
#mising values of 'umpire1' and 'umpire2' column replace with 'No_umpire' coz, that is missing
ipl['umpire1']=np.where((ipl['umpire1'].isnull()),"no_umpire1",ipl['umpire1'])
ipl['umpire2']=np.where((ipl['umpire2'].isnull()),"no_umpire2",ipl['umpire2'])
```
